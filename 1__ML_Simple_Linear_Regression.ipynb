## Coefficient of determination (R2)

```python
r2_score = corr **2  # not: bu formül sadece simple linear regressiom için geçelidir.
r2_score             # iki değerin biribiryle ne kadar ifade ettiğini anlamlı olduğunu verir.
                    # target'ı doğru tahmin etmeme için elimde ne kadar yetrli veri var. elimdeki ver %75 oranında target'ı tahmin için yeterlidir.
# modelimizin başarılı olup olmadığını ölçen metriklerden bir tanesidir==> R2 Score    
```

## Linear Regression

```python
sns.regplot(x="total_spend", y= "sales", data=df, ci=None) # x ve y nin regression grafiğini çizdiriyoruz best fit line olarak veriyor.
plt.show()

#not: best fit line'ın altında ve üstünde kalan değerleri topladığımızda 0 çıkar. 
# bu nedenle gerçek değerle tahmini değer arasında kalan 
# farkın karesini aldığımızda minimum değeri verir best line çizgiyi çizer
```

## Splitting the dataset into X(independent variables) and y (dependent variable)

```python
# y_pred = b1X + b0   best fit line çizdikten sonra formülle  y'yi tahmin edeceğiz. 
```

```python
X = df["total_spend"] # x'i bağımsız, y'yi de bağımlı değişke olarak böldük 
y = df["sales"]
```

## Determination of coefficients (slope and intercept)

```python
np.polyfit(X, y, deg=1) # scikit learn ile yapacağız
```

```python
slope, intercept = np.polyfit(X, y , deg=1)
print("slope    :", slope)   # line'ın eğimi
print("intercept:", intercept) #x = 0 iken  y nin ilk başlangıç değeri
```

## Why do we use the least squares error method to find the regression line that best fits the data?

```python
# y_pred = b1X + b0
```

```python
b1, b0 = np.polyfit(X, y , deg=1)
```

```python
print("b1 :",b1)
print("b0 :", b0)
```

```python
y_pred = b1*X + b0   
```

```python
values = {"actual" :y, "predicted": y_pred, "resudial": y- y_pred, "LSE": ( y- y_pred)**2}
df_2 = pd.DataFrame(values)
df_2

#LSE = Least Squares Error
```

```python
df_2.resudial.sum().round() # LSE karelerini almamıızn nedeni: 
```

```python
df_2.LSE.sum()
```

## Prediction with simple linear regression

```python
patential_spend = np.linspace(0,500,100) # potansiyel harcama için değerler alıyorum yani x'im bu değer oldu
patential_spend
```

```python
predicted_sales_lin = b1 * patential_spend + b0
predicted_sales_lin  # y_pred olarak bu formülden gelep değerleri verip aşağıda grafikte tahminimizi çizdiriyoruz.
```

```python
plt.plot(patential_spend, predicted_sales_lin)
sns.scatterplot(x ="total_spend", y= "sales", data=df)
plt.show()
```
