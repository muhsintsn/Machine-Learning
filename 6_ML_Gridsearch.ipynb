#  Gridsearch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False)

poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

## Scaling the Data

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scaler = StandardScaler()    # target e kesinlikle scale uygulanmaz
```

```python
scaler.fit(X_train) # Önemli
```

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test) # önemli X_trainden aldığımız değerlere istinaden  X_test scale ediliyor 
X_test_scaled #sızıntıyı önledik ve iki tarafıda standarlaştırdık
```

```python
pd.DataFrame(X_train_scaled).agg(["mean", "std"]).round()
```

```python
pd.DataFrame(X_test_scaled).agg(["mean", "std"]).round()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, y_train)
```

```python
y_pred = lm.predict(X_test_scaled)
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
def color_red(val):                      # google 
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV   
# en iyi parmetre değeri onu verir, overfit varsa düzeltir. ML de bütün modellerde kullanılır. genel
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42)
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1], # hyperparametre. sınır değerler çıkınca yeni bir range verilebilir aplha nın sınırı yok istediğin kadar verebilirsin 0 olmaz
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]} # hyperparametre lasso oranı 0 ile 1 arasında olur
# dict içinde liste veririz.
#alpha, l1_ratio gibi isimler GrideSearch'ün içinde geçen parametre isimleri ile aynı olacak 
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2) # modeli, parametre grid'i score'u, cv yi verdik bu değerler üzerinden iterasyon yapacak bize en iyi parametre değrini verecek 
```

```python
grid_model.fit(X_train_scaled, y_train) #eğitimi yaptırıyoruz
```

```python
grid_model.best_params_ # en iyi değerler
```

```python
pd.DataFrame(grid_model.cv_results_) # score
```

```python
grid_model.best_index_ # 6. index en iyi rmse skorunun olduğu index
```

```python
grid_model.best_score_
```

## Using Best Hyper Parameters From GridSearch

```python
#grid_model == ElastcNet8alpha = 0.01, l1_ratio=1
y_pred = grid_model.predict(X_test_scaled)
y_train_pred = grid_model.predict(X_train_scaled)
# gride_model model gibi veririz yukarıda yaptığımız best_score değerleri içinde barındırıyor. 
```

```python
train_val(y_train, y_train_pred, y_test, y_pred, "GridSearch")
```
