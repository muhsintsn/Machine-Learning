Verileri önceden işleyin Eğitim ve değerlendirme için verileri bölün.

```python
def preprocess(data):
    X =data.iloc[:, 1:17] # tüm satırlar, tüm özellikler ve label yok
    
    y= data.iloc[:,17] # tüm satırlar, yalnızca etiket
    
    return X, y
```

```python
# Veri kümesini karıştır ve böl
# Scikit-learn sayesinde artık bunu kullanmamıza gerek yok! uzum metod
data = data.sample(frac=1).reset_index(drop=True)
data_total_len = data[data.columns[0]].size
data_train_frac = 0.9
split_index = math.floor(data_total_len*data_train_frac)
train_data = data.iloc[:split_index]
eval_data = data.iloc[split_index:]
```

Verileri bunun yerine daha az satır kullanarak scikit-learn kullanarak bölebiliriz.

```python
from sklearn.model_selection import train_test_split
```

```python
all_X, all_y = preprocess(data)
X_train, X_test, y_train, y_test = train_test_split(all_X, all_y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Modeli Eğitin ve Değerlendirin 

```python
 # Support Vector Classification. (Destek Vektör Sınıflandırması.)

clf = svm.SVC()
clf.fit(X_train, y_train)  # ????????
```

```python
clf.score(X_test, y_test)
```
